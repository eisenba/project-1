```python
import requests
import pandas as pd
import numpy as np
import json
```

```python
# cleaning Dallas 2018
dallas2018 = pd.read_csv('working/Dallas2018.csv')
# rename columns
dallas2018.columns = ['Group', 'Class', 'GradCount', 'GradRate', 'ContCount', 
                      'ContRate','TxCHSECount','TxCHSERate','DropCount',
                      'DropRate', 'NotDropCount', 'NotDropRate']
# remove top labels read as rows
dallas2018 = dallas2018.iloc[3:19]
# remove blank rows
dallas2018 = dallas2018.iloc[:-2:2]
# select grad rate
dallas2018 = dallas2018[['GradRate']]
# transpose table
dallas2018 = dallas2018.transpose()
dallas2018.columns = ['BLKGradRate', 'INDGradRate', 'ASNGradRate',
                      'LATGradRate', 'HIPGradRate','WHTGradRate', 'BIRGradRate']
dallas2018['District'] = 'Dallas Independent School District'
dallas2018
```

```python
# cleaning Houston 2018
houston2018 = pd.read_csv('working/Houston2018.csv')
# rename columns
houston2018.columns = ['Group', 'Class', 'GradCount', 'GradRate', 'ContCount', 
                      'ContRate','TxCHSECount','TxCHSERate','DropCount',
                      'DropRate', 'NotDropCount', 'NotDropRate']
# remove top labels read as rows
houston2018 = houston2018.iloc[3:19]
# remove blank rows
houston2018 = houston2018.iloc[:-2:2]
# select grad rate
houston2018 = houston2018[['GradRate']]
# transpose table
houston2018 = houston2018.transpose()
houston2018.columns = ['BLKGradRate', 'INDGradRate', 'ASNGradRate',
                      'LATGradRate', 'HIPGradRate','WHTGradRate', 'BIRGradRate']
houston2018['District'] = 'Houston Independent School District'
houston2018
```

```python
# cleaning San Antonio 2018
sanantonio2018 = pd.read_csv('working/SanAntonio2018.csv')
# rename columns
sanantonio2018.columns = ['Group', 'Class', 'GradCount', 'GradRate', 'ContCount', 
                      'ContRate','TxCHSECount','TxCHSERate','DropCount',
                      'DropRate', 'NotDropCount', 'NotDropRate']
# remove top labels read as rows
sanantonio2018 = sanantonio2018.iloc[3:19]
# remove blank rows
sanantonio2018 = sanantonio2018.iloc[:-2:2]
# select grad rate
sanantonio2018 = sanantonio2018[['GradRate']]
# transpose table
sanantonio2018 = sanantonio2018.transpose()
sanantonio2018.columns = ['BLKGradRate', 'INDGradRate', 'ASNGradRate',
                      'LATGradRate', 'HIPGradRate','WHTGradRate', 'BIRGradRate']
sanantonio2018['District'] = 'San Antonio Independent School District'
sanantonio2018
```

```python
TX2018 = pd.concat([dallas2018, houston2018, sanantonio2018])
TX2018 = TX2018.reset_index()
TX2018 = TX2018[['District','BLKGradRate', 'INDGradRate', 'ASNGradRate',
                      'LATGradRate', 'HIPGradRate','WHTGradRate', 'BIRGradRate']]

TX2018.to_csv('output/TX2018.csv')
TX2018
```
