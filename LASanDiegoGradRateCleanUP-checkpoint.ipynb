```python
import pandas as pd
```

```python
filename="c:\\users\\asimg\documents\\cohort1617.txt"
df = pd.read_csv(filename, sep='\t', lineterminator='\r')
```

```python
df.head()
```

```python
counties = ["Los Angeles", "San Diego"]
df_LA = df[df["CountyName"].isin(counties)]
df_LA = df_LA[df_LA["AggregateLevel"] == "D"]
```

```python
df_LA.head()
```

```python
df_LA = df_LA[df_LA["CharterSchool"]=="All"]
```

```python
keep_columns = ["AcademicYear", "CountyCode", "CountyName", "ReportingCategory", "CohortStudents","Regular HS Diploma Graduates (Count)", "Regular HS Diploma Graduates (Rate)" ]
df_LA = df_LA[keep_columns]                
```

```python
df_LA["CohortStudents"] = pd.to_numeric(df_LA["CohortStudents"], errors='coerce')
df_LA["Regular HS Diploma Graduates (Count)"] = pd.to_numeric(df_LA["Regular HS Diploma Graduates (Count)"], errors='coerce')
```

```python
df_final = df_LA.groupby(["AcademicYear","CountyName","ReportingCategory"])["CohortStudents", "Regular HS Diploma Graduates (Count)"].sum()

#df_LA.groupby(["AcademicYear","CountyName"])["CohortStudents"].sum()
```

```python
df_final = df_final.reset_index()
```

```python
df_final["Regular HS Diploma Graduates (Rate)"] = (df_final["Regular HS Diploma Graduates (Count)"] / df_final["CohortStudents"] ) *100
```

```python
df_final.loc[df_final['ReportingCategory'] == "RB", 'RACE/ETHNICITY'] = 'African American'
df_final.loc[df_final['ReportingCategory'] == "RI", 'RACE/ETHNICITY'] = 'American Indian or Alaska Native'
df_final.loc[df_final['ReportingCategory'] == "RA", 'RACE/ETHNICITY'] = 'Asian'
df_final.loc[df_final['ReportingCategory'] == "RF", 'RACE/ETHNICITY'] = 'Filipino'
df_final.loc[df_final['ReportingCategory'] == "RH", 'RACE/ETHNICITY'] = 'Hispanic or Latino'
df_final.loc[df_final['ReportingCategory'] == "RD", 'RACE/ETHNICITY'] = 'Not Reported'
df_final.loc[df_final['ReportingCategory'] == "RP", 'RACE/ETHNICITY'] = 'Pacific Islander'
df_final.loc[df_final['ReportingCategory'] == "RT", 'RACE/ETHNICITY'] = 'Biracial'
df_final.loc[df_final['ReportingCategory'] == "RW", 'RACE/ETHNICITY'] = 'White'
```

```python
df_final=df_final.dropna()
```

```python
df_final.to_csv("c:\\users\\asimg\Documents\\LA1617_GraduationFinalNew.csv")
```
